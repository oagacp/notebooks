# <center>Testing the Central limit theorem using simulation</center>

The idea is to demostrate using a simulation the central limit theorem. I will use a uniform, binomial and exponencial distribution.

The central limit theorem indicate for a random samples, each of size n, are taken from any population with mean u and stardard deviation s, the sampling distribution of the sample means (averages) will:

1. Have a mean 
2. Have a standard deviation (Standard error)
3. be approximately a normally distributed regardless of the shape of the parent population (normality improves with large n)

### 1.- Testing using Uniform distribution

```python
import numpy as np
```

First we will demostrate the theorem using the uniform distribution

```python
values_uniform = np.random.uniform(low = 0.0, high = 1.0, size=1000)
```

```python
import matplotlib.pyplot as plt
```

```python
count, bins, ignored = plt.hist(values_uniform, 15, density=True)

plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')

plt.show()
```

```python
Arr_sum = []
for i in range(0, 5, 1):
    sample = np.random.choice(values_uniform , size = 2, replace = False)
    mean_dist = np.mean(sample)
    Arr_sum = np.append(Arr_sum, mean_dist)
```

```python
import seaborn as sns
```

```python
sns.distplot(Arr_sum, hist=False)
plt.show() 
```

```python
Arr_sum1 = []
w = 20
for i in range(0, 1000, 1):
    sample = np.random.choice(values_uniform , size = w, replace = False)
    mean_dist = np.mean(sample)
    Arr_sum1 = np.append(Arr_sum1, mean_dist)
```

```python
sns.distplot(Arr_sum1, hist=False)
plt.show() 
```

```python
import scipy.stats as stats
```

```python
stats.shapiro(Arr_sum)
```

```python
stats.shapiro(Arr_sum1)
```

As p-value in both cases of shapiro test p-value > 0.05, where 0.05 is the cut off, there is enough evidence to not reject the null hypothesis, so the distribution is normal for both arrays.

### 2.- Testing using Binomial distribution

Now we will test the theorem using the binomial distribution

```python
n, p = 50, .3  # number of trials, probability of each trial
values_binomial = np.random.binomial(n, p, size=1000)
```

```python
bin = np.arange(0,30,1)
plt.hist(values_binomial, bins=bin, edgecolor='blue') 
plt.title("Binomial Distribution") 
plt.show()
```

```python
Arr_sum2 = []
for i in range(0, 1000, 1):
    sample = np.random.choice(values_binomial , size = 2, replace = False)
    mean_dist = (sample[0] + sample[1])/2
    Arr_sum2 = np.append(Arr_sum2, mean_dist)
```

```python
sns.distplot(Arr_sum2, hist=False)
plt.show() 
```

```python
Arr_sum3 = []
w = 20
for i in range(0, 1000, 1):
    sample = np.random.choice(values_binomial , size = w, replace = False)
    mean_dist = np.mean(sample)
    Arr_sum3 = np.append(Arr_sum3, mean_dist)
```

```python
sns.distplot(Arr_sum3, hist=False)
plt.show() 
```

```python
stats.shapiro(Arr_sum2)
```

```python
stats.shapiro(Arr_sum3)
```

The first shapiro test p-value < 0.05, we reject the null hypothesis, so the distribution of the mean using 2 values is not a normal distribution. For the second case, where to calcuate the mean we used 20 values p-value is > 0.05, where 0.05 is the cut off, there is enough evidence to not reject the null hypothesis, so the distribution is normal.

### 3.- Testing using Exponential distribution

Now is the turn to verify the theorem using exponential distribution

```python
values_exponential = np.random.exponential(scale=1.0, size=1000)
```

```python
bin = np.arange(0,10,0.1)

plt.hist(values_exponential, bins=bin, edgecolor='blue') 
plt.title("Exponential Distribution") 
plt.show()
```

```python
Arr_sum4 = []
for i in range(0, 1000, 1):
    sample = np.random.choice(values_exponential , size = 2, replace = False)
    mean_dist = (sample[0] + sample[1])/2
    Arr_sum4 = np.append(Arr_sum4, mean_dist)
```

```python
sns.distplot(Arr_sum4, hist=False)
plt.show()
```

```python
Arr_sum5 = []
w = 80
for i in range(0, 1000, 1):
    sample = np.random.choice(values_exponential , size = w, replace = False)
    mean_dist = np.mean(sample)
    Arr_sum5 = np.append(Arr_sum5, mean_dist)
```

```python
sns.distplot(Arr_sum5, hist=False)
plt.show()
```

```python
stats.shapiro(Arr_sum4)
```

```python
stats.shapiro(Arr_sum5)
```

The first shapiro test p-value < 0.05, we reject the null hypothesis, so the distribution of the mean using 2 values from an exponential distribution is not a normal distribution. For the second case, where to calcuate the mean we used 80 values p-value is > 0.05, where 0.05 is the cut off, there is enough evidence to not reject the null hypothesis, so the distribution is normal.
